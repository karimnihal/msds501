```python
import matplotlib.pyplot as plt

plt.figure(figsize=(5, 3))
im = plt.imread("cat_istanbul.jpg")
plt.imshow(im) 
plt.show() 
```

```python
im_gray = im[:,:,0]
im_gray.shape
```

```python
plt.figure(figsize=(5, 3))
plt.imshow(im_gray, cmap='gray') 
plt.show()
```

```python
def show_images(image1, image2, title1='Image 1', title2='Image 2'):
    fig, axes = plt.subplots(1, 2, figsize=(9, 5))

    axes[0].imshow(image1, cmap='gray')
    axes[0].set_title(title1)
    axes[0].axis('off')

    axes[1].imshow(image2, cmap='gray')
    axes[1].set_title(title2)
    axes[1].axis('off')

    plt.show()
```

```python
im_noisy = plt.imread("noisy_cat.jpg")
show_images(im_gray, im_noisy, title1='Original', title2='Noisy')
```

## Flip

```python
im_flipped = horizontal_flip(im_gray)
```

```python
show_images(im_gray, im_flipped, title1='Original', title2='Flipped')
```

## Blur

```python
im_blurred = blur(im_gray, 9)
```

```python
show_images(im_gray, im_blurred, title1='Original', title2='Blurred')
```

## Denoise

```python
im_denoised = denoise(im_noisy)
```

```python
show_images(im_denoised, im_noisy, title1='Denoised', title2='noisy')
```

## Thresholding 

```python
im_th = thresholding(im_gray)
```

```python
show_images(im_gray, im_th, title1='Original', title2='Thresholded')
```

## Center Crop

```python
im_cropped = center_crop(im_gray, x_size=400, y_size=600)
```

```python
im_gray.shape, im_cropped.shape
```

```python
show_images(im_gray, im_cropped, title1='Original', title2='Cropped')
```
